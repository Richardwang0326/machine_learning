```python
#library
import numpy as np
import matplotlib.pyplot as plt
import csv

lnchprg=[]
math10=[]

# load csv file
with open ('MEAP93.csv', newline = '') as csvfile:
    
    # stored as dict
    rows = csv.DictReader(csvfile)
    
    # load need data
    for row in rows:
        lnchprg.append((float)(row['lnchprg']))
        math10.append((float)(row['math10']))

# calculate mean
lnchprg_mean = np.mean(np.array(lnchprg))
math10_mean = np.mean(np.array(math10))

# calculate beta0 and beta1
fraction = 0
denominator = 0

for i in range(len(lnchprg)):
    fraction += (lnchprg[i]-lnchprg_mean)*(math10[i]-math10_mean)
    denominator += (lnchprg[i]-lnchprg_mean)**2
    
beta1 = fraction / denominator
beta0 = math10_mean - beta1 * lnchprg_mean

print(beta1,beta0)
```

```python
# draw line based on OLS
def graph(formula, x_range):
    x = np.array(x_range)
    y = eval(formula)
    plt.plot(x,y,'r')
# plotting the prediction line
graph('beta0 + beta1 * x', range(90))
x = np.array(lnchprg)
y = np.array(math10)
plt.xlabel('lnchprg')
plt.ylabel('math10')
plt.scatter(x,y)
```
